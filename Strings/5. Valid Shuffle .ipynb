**Question 5 **
Check if a string is a valid shuffle of two other strings
You are given 3 strings:
first
second
result

**Example - **
first  = "XY"
second = "12"

Valid shuffles:
"X1Y2"
"1X2Y"
"1XY2"
"Y1X2"

‚ùå Invalid: "Y21XX"  (extra X, wrong count)

üîÑ What does ‚Äúshuffle‚Äù mean here?

Shuffle means interleaving, NOT random mixing.

You can:

1. Take all characters from first

2. Take all characters from second

3. Mix them together (But relative order inside each string must stay the same)

There are TWO DIFFERENT PROBLEMS that look similar:

üîπ Problem 1 (LOOSE shuffle ‚Äì your Java example)

‚úî Uses sorting
‚úî Order ignored
‚úî 21XY is valid

üîπ Problem 2 (STRICT shuffle ‚Äì common interview version)

‚ùå Sorting NOT enough
‚úî Order must be preserved
‚úî 21XY is invalid

```python
# Problem - 1 : Loose Shuffle
def is_valid_shuffle(first, second, result):
    if len(first) + len(second) != len(result):
        return False

    return sorted(first + second) == sorted(result)
print(is_valid_shuffle("XY", "12", "1XY2"))   # True
print(is_valid_shuffle("XY", "12", "Y1X2"))   # True
print(is_valid_shuffle("XY", "12", "Y21XX"))  # False
print(is_valid_shuffle("XY", "12", "XY1"))    # False
print(is_valid_shuffle("XY", "12", "YX21"))    # False in case of Strict shuffle and True in case of Loose shuffle
```

```python
# Problem - 2 : Strict Shuffle
def is_strict_shuffle(first, second, result):
  if len(first) + len(second) != len(result):
    return False
  i = j = 0
  for ch in result:
    if i < len(first) and ch == first[i]:
      i += 1
    elif j < len(second) and ch == second[j]:
      j += 1
    else:
      return False
  return True
print(is_strict_shuffle("XY", "12", "1XY2"))   # True
print(is_strict_shuffle("XY", "12", "Y1X2"))   # False in case of Strict shuffle and True in case of Loose shuffle
print(is_strict_shuffle("XY", "12", "Y21XX"))  # False
print(is_strict_shuffle("XY", "12", "XY1"))    # False
print(is_strict_shuffle("XY", "12", "YX21"))   # False in case of Strict shuffle and True in case of Loose shuffle
```
