🔹**Problem Understanding** (Simple Words)

You are given a string s.

👉 You need to find the length of the longest substring

👉 such that no character repeats

Important:

Substring = continuous part

Not allowed: skipping characters

🔹 **Example**

Input:
s = "abcabcbb"

Possible substrings without repeating:

"abc" ✅ length = 3

"bca" ✅ length = 3

"cab" ✅ length = 3

👉 Answer = 3

🔁 Visual Flow

**For "abba":**

i = 0

Start:

seen = {}

j = 0 → 'a'

check → not present

add → {'a'}

j = 1 → 'b'

check → not present

add → {'a','b'}

j = 2 → 'b'

check → already present ❌ break

Submission Link - https://leetcode.com/problems/longest-substring-without-repeating-characters/submissions/1966061257/

```python
s = "abcabcd"
def longest_substring(s):
  max_length = 0
  for i in range(len(s)):
    count = 0
    seen = set()
    for j in range(i,len(s)):
      if s[j] in seen:
        break
      count += 1
      seen.add(s[j])
    max_length = max(max_length,count)
  return max_length
longest_substring(s)
```
