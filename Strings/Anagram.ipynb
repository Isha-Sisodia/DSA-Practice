🔹 Simple Definition of **Anagram**

👉 If two words have the same letters in different order, they are anagrams.

🔹 **Examples**
listen → silent
evil → vile
dusty → study
triangle → integral

🔹**Why Dictionary is Perfect Here**

Dictionary gives:
👉 Key → Group mapping

Example:

{
  
  "aet": ["eat", "tea", "ate"],

  "ant": ["tan", "nat"],

  "abt": ["bat"]

}

👉 Here:

Key = sorted word

Value = list of anagrams

**Step 1:** Group using dictionary

anagram_map = {

⚡ Key Insight

👉 Dictionary is used for building the groups efficiently
👉 List of lists is just the final format

We use a dictionary to group anagrams.
For every word, we sort its letters and store it as a string called key.

Example:
"eat" → "aet"
"tea" → "aet"

Since both have the same key, they belong to the same group.

For each word:

Create key = "".join(sorted(word))

Check if this key exists in the dictionary

If not, create an empty list for it
Add the word to that list

The dictionary stores:
sorted word → list of original words

At the end, we return all the values (groups).

Important:

- key is just a string, not a list
- Checking key in dict only checks dictionary keys
- Same key means same anagram group
- Words are never overwritten, only appended

```python
strs = ["eat", "tea", "tan", "ate", "nat", "bat", "aet"]
def group_anagram(strs):
  anagram_dict = {}    # dictionary to store groups
  for word in strs:
    key = "".join(sorted(word))      # STEP 1: create key string (sorted word)
    if key not in anagram_dict:      # if that key string is not present as a key in the dictinary
      anagram_dict[key] = []         # then it creates an empty list as the value in the dictionary
    anagram_dict[key].append(word)   # now dict ki values mein vo word append ho jayega jisko humne sort kiya tha
  return list(anagram_dict.values())
group_anagram(strs)
```

```python
# submitted code on leedcode
class Solution(object):
    def groupAnagrams(self, strs):
        dict = {}
        for word in strs:
            var = "".join(sorted(word))
            if var not in dict:
                dict[var] = []
            dict[var].append(word)
        return list(dict.values())
```

there is a way to solve this without using a dictionary and It works, but it’s less efficient and not preferred in interviews.

💡 Idea:

- Maintain a result list
- For each word, check existing groups manually
- If match found → add there
- Else → create new group

```python
# This is the code using List
def groupAnagrams(strs):
    result = []

    for word in strs:
        placed = False

        for group in result:
            # compare with first word of group
            if sorted(group[0]) == sorted(word):
                group.append(word)
                placed = True
                break

        if not placed:
            result.append([word])

    return result


# Test
print(groupAnagrams(["eat","tea","tan","ate","nat","bat"]))
```
