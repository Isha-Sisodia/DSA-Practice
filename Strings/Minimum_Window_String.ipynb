```python
def minWindow(s, t):
    t_dict = {}
    for ch in t:
        if ch in t_dict:
            t_dict[ch] += 1
        else:
            t_dict[ch] = 1

    window_dict = {}

    have = 0
    need = len(t_dict)

    left = 0
    min_len = float("inf")
    result = ""

    for right in range(len(s)):
        ch = s[right]

        # add character
        if ch in window_dict:
            window_dict[ch] += 1
        else:
            window_dict[ch] = 1

        # check match
        if ch in t_dict and window_dict[ch] == t_dict[ch]:
            have += 1

        # shrink window
        while have == need:
            # update result
            # if (right - left + 1) < min_len:
            if len(s[left:right+1]) < min_len:
                # min_len = right - left + 1
                min_len = len(s[left:right+1])
                result = s[left:right+1]

            # remove left char
            window_dict[s[left]] -= 1
            if s[left] in t_dict and window_dict[s[left]] < t_dict[s[left]]:     #whether have would be upadated or remains same!
                have -= 1

            left += 1

    return result
```
